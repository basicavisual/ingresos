```python
import pandas as pd
import csv

file = 'resultado.csv'

ingresos = 'ingresos_comp.csv'
egresos = 'Egresos_comp.csv'

ingresos_csv = pd.read_csv(ingresos, encoding='Latin-1')
egresos_csv = pd.read_csv(egresos, encoding='Latin-1')

ingresos_df = pd.DataFrame(ingresos_csv)
egresos_df = pd.DataFrame(egresos_csv)

egresos_df
```

```python
with open(file, 'a') as csvFile:
    for index, row in egresos_df.iterrows():
        
        fuente_egresos = row['Fuente Concat']        
        finalidad = row['Finalidad']
        unidad_responsable = row['Unidad Responsable']
        monto_ejercido = row['Unidad Responsable']
        
        row_egresos = [fuente_ingresos, finalidad, unidad_responsable, monto_ejercido]


        for index, row in ingresos_df.iterrows():
            fuente_ingresos = row['FUENTE_CONCAT']
            if fuente_egresos == fuente_ingresos:
                
                row_egresos.insert(1, finalidad)
                row_egresos.insert(1, unidad_responsable)
                row_egresos.insert(1, monto_ejercido)
                
                writer = csv.writer(csvFile)
                writer.writerow(row_egresos)
csvFile.close()
```
